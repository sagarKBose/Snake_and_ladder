```python
import random                   #Importing the random library
```

```python
#Creating my custom class color to add color in between output texts

class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
#Given, Snake and Ladder dictionary{key:value} containing "key" as the position of snakes/ladder 
#and the "value" which denotes the position where it will take the player

snakes = {17 : 7, 54 : 34, 62 : 19, 98 : 79}
ladders = {3 : 38, 24 : 33, 42 : 93, 72 : 84}

goal = 100
dice = 6
```

```python
#This function reads the names of the 2 player from the users and returns to the auto_mode()

def getting_player_names():
    player1_name = None
    while not player1_name:
        player1_name = input("Enter Player1's Name   ").strip()
        
    player2_name = None
    while not player2_name:
        player2_name = input("Enter Player2's Name   ").strip()
    
    print("Game will be played between Player1: "+ color.BOLD+color.BLUE+player1_name+color.END+ " and Player2: "+color.BOLD+color.GREEN+player2_name+color.END)
    return player1_name,player2_name
```

```python
#This function gets the dice value and returns to the auto_mode()

def getting_dice_value():
    dice_value = random.randint(1,dice)
    print("Dice Value is "+color.BOLD+color.YELLOW+str(dice_value)+color.END)
    return dice_value
```

```python
#This function checks the current value of the palyer with
#the goal(100), whether there is any snake bite or not
#or the player got any ladder or not
#and finally returns the resultant value of the player

def snake_and_ladder(dice_value, player_name, curr_value):
    old_value = curr_value
    curr_value = curr_value + dice_value
    
    if(curr_value>goal):  #if the current value of the player is more than 100, then let the player keep on trying
        print("\nYou need to get exactly "+ str(goal-old_value)+" to reach your goal. Keep trying your luck")
        return old_value
    
    
    print("\n"+player_name+" moved from "+str(old_value)+" to "+str(curr_value))
    
    if curr_value in snakes: #if the current value is snake's head
        print("\n"+color.BOLD+color.UNDERLINE+"Oops, you got a"+color.RED+" SNAKE BITE!"+color.END)
        final_value = snakes.get(curr_value)
        print("\n"+str(player_name)+" was moved DOWN from "+str(curr_value)+" to "+ str(final_value))
      
    
    elif curr_value in ladders:     #if the current value is a ladder
        print("\n"+color.BOLD+color.UNDERLINE+"Yayy, you got a"+color.RED+" LADDER!"+color.END)
        final_value = ladders.get(curr_value)
        print("\n"+str(player_name)+" was moved UP from "+str(curr_value)+" to "+ str(final_value))
        
    else:
        final_value = curr_value
    return final_value
```

```python
#Checking the winner 

def checking_winner(curr_val,player_name):
    if curr_val==goal:
        print("\nHURRAH, "+color.BOLD+player_name +color.END+" IS THE WINNER!!")
        print("Thank you for playing the game, have a good day. :)")
        #sys.exit(1)
        return True
```

```python
#Function for the user to choose Auto or Manual Mode

def choose_auto_or_manual():
    mode = None
    while not mode:
        print("\n\nKindly choose the MODE in which you want to play the game ")
        mode = input('Press '+ color.BOLD+ '"A" '+color.END + 'for Auto-Mode or Press'+color.BOLD+ ' "M" '+color.END +'for Manual-Mode:   ').lower()
        
        while (mode!='a' and mode!='m'):       #If the user enters something other than 'm' or 'n'
            print("Kindly Choose The MODE in which you want to play the game ")
            mode = input('Press '+ color.BOLD+ '"A" '+color.END + 'for Auto-Mode or Press'+color.BOLD+ ' "M" '+color.END +'for Manual-Mode:   ').lower()
    return mode
```

```python
#Function for MANUAL mode

def manual_mode():
        print("""\n Getting Familiar with the Rules:
        1. The game will be played between 2 players.
        2. The player starts from '0' and has to reach the final position that is '100'.
        3. Each player will get a choice to input a DESIRED VALUE between 1 and 20. 
           Move ahead as per the outcome of your choice.
        4. There might be 'SNAKES' on your way which might bring you down.
        5. There might be 'LADDERS' on your way which might bring you up.
        6. You can quit the game anytime by typing "99"
        7. Enjoy playing the game!
    """)
        
        player1_name,player2_name = getting_player_names()

        
        #Intitial postions of player1 and player2 are 0
        player1_curr_pos = 0
        player2_curr_pos = 0
        quit_status = False
        while True:
            
           #For Player1 
           #--------------# 
            print("\nYou can quit anytime. Just type"+color.BOLD+ "'99' to quit."+color.END)
            input1 = int (input("\n"+color.BLUE+color.BOLD+player1_name+color.END+",Please Enter a number between 1-20:   "))
            
            if input1==99:        #Press 99 to quit the game
                print("\nThank you for playing. \nExiting the Game!")
                break
                
            while(input1<1 or input1>20):     #If the user gives a wrong input
                print("OOPs, you entered a wrong entry! Please Enter a number between 1-20")
                print("\nYou can quit anytime. Just type"+color.BOLD+ "'99' to quit."+color.END)
                
                input1 = int (input("\n"+color.BLUE+color.BOLD+player1_name+color.END+",Please Enter a number between 1-20:   "))
                
                if input1==99:    #Press 99 to quit the game
                    print("\nThank you for playing. \nExiting the Game!")
                    quit_status = True
                    break
            
            if quit_status == True:
                break
            
            print("\nMoving Ahead...")        

            print(f"Player1: {player1_name} is moving... ")

            player1_curr_pos = snake_and_ladder(input1,player1_name,player1_curr_pos)  #Checking whether it's a snake or ladder

            if checking_winner(player1_curr_pos,player1_name) == True:   #If WINNER, then break from the while loop to end the game
                break
            
            #---------------------------------------------------------------#
            #Repeating the above steps for player2
            #For Player2 
            #--------------# 
            print("\nYou can quit anytime. Just type"+color.BOLD+ "'99' to quit."+color.END)
            input2 = int (input("\n"+color.BLUE+color.BOLD+player2_name+color.END+",Please Enter a number between 1-20:   "))
            
            if input2==99:        #Press 99 to quit the game
                print("\nThank you for playing. \nExiting the Game!")
                break
                
            while(input2<1 or input2>20):     #If the user gives a wrong input
                print("OOPs, you entered a wrong entry! Please Enter a number between 1-20")
                print("\nYou can quit anytime. Just type"+color.BOLD+ "'99' to quit."+color.END)
                input2 = int (input("\n"+color.BLUE+color.BOLD+player2_name+color.END+",Please Enter a number between 1-20:   "))
                
                if input2==99:    #Press 99 to quit the game
                    print("\nThank you for playing. \nExiting the Game!")
                    quit_status = True
                    break
            
            if quit_status == True:
                break
            
            print("\nMoving Ahead...")

            print(f"Player2: {player2_name} is moving... ")

            player2_curr_pos = snake_and_ladder(input2,player2_name,player2_curr_pos)  #Checking whether it's a snake or ladder

            if checking_winner(player2_curr_pos,player2_name)==True:   #If WINNER, then break from the while loop to end the game
                break
            
```

```python
#Function for AUTO mode

def auto_mode():
        print("""\n Getting Familiar with the Rules:
        1. The game will be played between 2 players.
        2. The player starts from '0' and has to reach the final position that is '100'.
        3. Each player will get a chance to roll the dice. Move ahead as per the outcome of your dice.
        4. There might be 'SNAKES' on your way which might bring you down.
        5. There might be 'LADDERS' on your way which might bring you up.
        6. You can quit the game anytime by typing "quit"
        7. Enjoy playing the game!
    """)
        
        player1_name,player2_name = getting_player_names()

        #Intitial postions of player1 and player2 are 0
        player1_curr_pos = 0
        player2_curr_pos = 0
        quit_status = False
        while True: 
             
            #For Player1 
            #--------------# 
            print("\nYou can quit anytime. Just type"+color.BOLD+ "'QUIT'"+color.END)
            input1 = input("\n"+color.BLUE+color.BOLD+player1_name+color.END+", please type"+ color.BOLD+" 'Roll'" +color.END+ " to get your dice value:   ").lower()
            
            if input1 == "quit":      #Press 'quit' to quit the game
                print("\nThank you for playing. \nExiting the Game!")
                break
                
            while(input1!="roll"):       #If the user gives a wrong input
                print("OOPs, you entered a wrong entry! Please type 'roll' to get your dice value")
                input1 = input("\n"+color.BLUE+color.BOLD+player1_name+color.END+", please type"+ color.BOLD+" 'Roll'" +color.END+ " to get your dice value:   ").lower()
                
                if input1 == "quit":      #Press 'quit' to quit the game
                    print("\nThank you for playing. \nExiting the Game!")
                    quit_status = True
                    break
                    
            if quit_status == True:
                break
            
            
            print("\nRolling Dice...")
            
            dice_value = getting_dice_value()   #Getting a random dice value

            print(f"\nPlayer1: {player1_name} is moving... ")

            player1_curr_pos = snake_and_ladder(dice_value,player1_name,player1_curr_pos) #Checking whether it's a snake or ladder

            if checking_winner(player1_curr_pos,player1_name) == True:  #If WINNER, then break from the while loop to end the game
                break
            
            
            #---------------------------------------------------------------#
             
           #For Player2 
           #--------------# 
            
            print("\nYou can quit anytime. Just type"+color.BOLD+ "'QUIT'"+color.END)
            input2 = input("\n"+color.GREEN+color.BOLD+player2_name+color.END+", please type"+color.BOLD+ " 'Roll'" +color.END+ " to get your dice value:   ").lower()
            
            if input2 == "quit":    #Press 'quit' to quit the game
                print("\nThank you for playing. \nExiting the Game!")
                break
                
            while(input2!="roll"):       #If the user gives a wrong input
                print("OOPs, you entered a wrong entry! Please type 'roll' to get your dice value")
                input2 = input("\n"+color.GREEN+color.BOLD+player2_name+color.END+", please type"+color.BOLD+ " 'Roll'" +color.END+ " to get your dice value:   ").lower()
                
                if input2 == "quit":      #Press 'quit' to quit the game
                    print("\nThank you for playing. \nExiting the Game!")
                    quit_status = True
                    break
                    
            if quit_status == True:
                break
                
                
            print("\nRolling Dice...")
            
            dice_value = getting_dice_value()   #Getting a random dice value

            print(f"\nPlayer2: {player2_name} is moving... ")

            player2_curr_pos = snake_and_ladder(dice_value,player2_name,player2_curr_pos)  #Checking whether it's a snake or ladder

            if checking_winner(player2_curr_pos,player2_name)==True: #If WINNER, then break from the while loop to end the game
                break
            
            
```

```python
#This funtion displays the welcome text.
#Allows user to choose MANUAL mode or AUTO mode

def start_the_game():
    
    text = color.BOLD+color.DARKCYAN+"""
    ###### Welcome to Snakes & Ladders Game ###### 
           Developed By- Sagar Krishna Bose       
    \n\n"""+color.END
    print(text)
    
    mode = choose_auto_or_manual()   #Allow the user to choose the mode 
    
    if mode=='a':
        auto_mode()
        
    elif mode=='m':
        manual_mode()
```

```python
#Run this cell to start the game
if __name__ == "__main__":
    start_the_game()
```
